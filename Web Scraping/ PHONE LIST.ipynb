```
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

array_nome = list()
array_Endereço = list()
array_Bairro = list()
array_Telefone = list()
array_Site = list()

c = 0


try:
    html = urlopen("https://www.guiadesampa.com.br/")
except HTTPError as e:
    print(e)
except URLError:
    print("Server down or incorrect domain")
else:
    res = BeautifulSoup(html.read(),"html5lib")
    tags = res.findAll("div", {"class": "linha"})
    for tag in tags:
        tgs = tag.findAll("li")
        txt = str(tgs)

        a = (txt.find("=", 18))+1
        b = (txt.find(">", 18))-1

        url = f"https://www.guiadesampa.com.br/depto.asp?depto={txt[a:b]}"
        html = urlopen(url)
        res = BeautifulSoup(html.read(),"html5lib")
        

        # PEGAR NOME DO LUGAR
        nomefantasia = res.findAll("div", {"class": "nomefantasia"})
        txt = str(nomefantasia)
        a = b = 0
        a = (txt.find("<strong>", a))+8
        b = txt.find("</strong>", b)
        array_nome.append(txt[a:b])


        # PEGAR ENDEREÇO DO LUGAR
        endereco2 = res.findAll("div", {"class": "endereco2"})
        txt = str(endereco2)
        a_endereco2 = b_endereco2 = 0
        a_endereco2 = (txt.find("</strong>", a_endereco2))+9
        b_endereco2 = txt.find("</p>", b_endereco2)
        array_Endereço.append(txt[a_endereco2:b_endereco2])


        # PEGAR BAIRRO DO LUGAR
        bairro = res.findAll("div", {"class": "bairro2"})
        txt = str(bairro)
        a_bairro = b_bairro = 0
        a_bairro = (txt.find("<p>"))+3
        b_bairro = txt.find("</p>", b_bairro)
        txt = txt[a_bairro:b_bairro].replace(" ", "").replace("\n", "")

        idx1 = txt.find("-SãoPaulo/SP")
        txt = txt[:idx1].replace("-", " ")
        array_Bairro.append(txt)


        # PEGAR TELEFONE DO LUGAR
        telefone2 = res.findAll("div", {"class": "telefone2"})
        txt = str(telefone2)
        a_telefone2 = b_telefone2 = 0
        a_telefone2 = (txt.find("</strong>"))+9
        b_telefone2 = txt.find("</p>", b_telefone2)
        txt = txt[a_telefone2:b_telefone2]
        array_Telefone.append(txt)



        # PEGAR TELEFONE DO LUGAR
        siteemail2 = res.findAll("div", {"class": "siteemail2"})
        txt = str(siteemail2)
        a_siteemail2 = b_siteemail2 = 0
        a_siteemail2 = (txt.find("false;"))+8
        b_siteemail2 = txt.find("</a>", b_siteemail2)
        txt = txt[a_siteemail2:b_siteemail2]
        array_Site.append(txt)
        if c > 48:
            break
        c += 1
dataset = pd.DataFrame({'Nome': array_nome,'Endereço': array_Endereço, 'Bairro': array_Bairro, 'Telefone': array_Telefone, 'Site': array_Site})

2222
```
